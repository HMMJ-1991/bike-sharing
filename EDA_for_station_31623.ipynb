### <font color="red">Import packages </font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('data/bike_sharing_data/daily_count.csv')
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data['daya of week']=data['date'].dt.weekday  #It is assumed the week starts on Monday, which is denoted by 0 and ends on Sunday which is denoted by 6
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
plt.xticks(rotation=45)
ax.scatter(data['date'][data['daya of week']==0],data['number of users'][data['daya of week']==0],color='#0048BA',label="Monday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==1],data['number of users'][data['daya of week']==1],color='#B284BE',label="Tuesday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==2],data['number of users'][data['daya of week']==2],color='#FFF600',label="Wenesday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==3],data['number of users'][data['daya of week']==3],color='#F4C2C2',label="Thursday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==4],data['number of users'][data['daya of week']==4],color='#7FFF00',label="friday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==5],data['number of users'][data['daya of week']==5],color='#FF0000',label="saturday",cmap=cm.brg)
ax.scatter(data['date'][data['daya of week']==6],data['number of users'][data['daya of week']==6],color='#000000',label="sunday",cmap=cm.brg)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Number of users')
plt.title('Scatter plot for daily demand')
plt.show()
```
